```python
from skimage.filters import sobel,scharr,prewitt,roberts
from skimage import morphology
from tqdm import tqdm_notebook as tqdm
from matplotlib import pyplot as plt
from sklearn.metrics import jaccard_similarity_score
import numpy as np
import cv2
import os
```

```python
DATA_PATH = 'Breast cancer images/Dataset'
```

```python
def create_test_data():
    testing_data = []
    for img in tqdm(os.listdir(DATA_PATH)):
        if img != "Thumbs.db" :
            path = os.path.join(DATA_PATH, img)
            img_data = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
            testing_data.append([np.array(img_data),int(img[:-4])])
    return testing_data
```

```python
data = create_test_data()
```

```python
def printImage(data):
    plt.imshow(data, cmap='gray',vmin=0,vmax=255)
    plt.show()
```

```python
def findSegmentation(data):
    gambar=data[0]
    squareDilated=morphology.disk(3)
    squareEroted=morphology.disk(6)
    
    
    printImage(gambar)
    
    elevation_map = scharr(gambar)

    print("Edge detection using : prewitt")
    
    
    print(elevation_map)
    
    plt.imshow(elevation_map, cmap='gray')
    plt.show()
    
    
    markers = np.zeros_like(gambar)
    
    markers[gambar < 50] = 255
    markers[gambar > 150] = 0
    
    cv2.imwrite("greyscale.png",markers)
    
    printImage(markers)
    
    dilated = morphology.dilation(markers,squareDilated)
    print("Dilasi")
    printImage(dilated)
    
    eroted = morphology.erosion(dilated,squareEroted)
    print("Erosi")
    printImage(eroted)
    

    
    print("Watershed")
    
    segmentation = morphology.watershed(elevation_map, eroted)

    printImage(segmentation)
    
    return segmentation
```

```python
def color(data):
    newData=[]
    for gambar in data:
        newGambar = findSegmentation(gambar)
        newData.append([newGambar,gambar[1]])
    return newData
```

```python
newData = color(data)
```

```python
cv2.imwrite("greyscale.png",data[0])
cv2.imwrite("segmentation.png",newData[0])
plt.imshow(newData[0], cmap='gray')
plt.show()
```
